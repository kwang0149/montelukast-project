# Dataset

### ✍️ INPUT

```python
dataset = pd.read_csv('out (all bogor 109ribuan).csv')
```

```python
provinces = pd.read_csv('provinces.csv')
cities = pd.read_csv('cities.csv')
districts = pd.read_csv('districts.csv')
sub_districts = pd.read_csv('sub_districts.csv')
```

```python
province = 'province'
city = 'city'
district = 'district'
subdistrict = 'subdistrict'
sub_district = 'sub_district'
postal_code = 'postal_code'
```

# Rename Attribute

```python
dataset.rename(columns={subdistrict: sub_district, 'geometry': 'location'}, inplace=True)
```

# Uppercasing

```python
dataset[province] = dataset[province].str.upper()
dataset[city] = dataset[city].str.upper()
dataset[district] = dataset[district].str.upper()
dataset[sub_district] = dataset[sub_district].str.upper()
```
